### (2)  메타문자 사용법 기본 내용

#### 0) 기본 축약표현식(참고: ^는 NOT의 의미)
- **숫자 표현식: [0-9] -> 축약:\d**
  - 숫자를 찾는데 사용


- **숫자 제외 표현식: [^0-9] -> 축약: \D**
  - 숫자를 제외한 모음 문자(텍스트+특수문자+화이트 스페이스) 찾는데 사용 (화이트 스페이스는 스페이스,tab, new line을 포함)
  

- **whitespace 표현식: [\t\n\r\f\v] -> 축약: \s**
  - whitespace 문자 찾는데 사용(스페이스,tab, new line)


- **whitespace 제외 표현식: [^\t\n\r\f\v] -> 축약: \S**
  - whitespace 문자가 아닌것(텍스트+특수문자+숫자) 찾는데 사용


- **문자+숫자 찾는 표현식: [a-zA-Z0-9_] -> 축약: \w**
  - 텍스트+숫자 찾는데 사용(단, 언더스코어는 포함)


- **문자+숫자 아닌 것 찾는 표현식: [^a-zA-Z0-9_] -> 축약: \W**
  - 특수문자+공백만 찾는데 사용

#### 1) [ ]
- 대괄호 안에는 어떤 것이든 들어갈 수 있음 (단, 대소문자 구분)
- 안에 들어가는 내용은 or(|)과 하이픈(-)으로 연결 가능(ex. [a-z]는 a부터 z를 의미)
- [abc]는 'a,b,c'중 1개 문자와 매칭

```python
# 연습
a=re.compile('a[a-z0-9]z')
# a=re.compile('a[a-z0-9]+z')
# a=re.compile('a[a-z0-9]*z')
result = re.search(a, 'a999z') 
print(result)
```

#### 2) Dot(.) 
- \n을 제외한 모든 문자와 매칭
- Dot(.) 하나(두개)는 문자 하나(두개)를 의미
  - 참고: 문자는 숫자(0-9)와 특수문자(!@#$%^& 등)를 모두 포함

```python
# 연습
a=re.compile('a.z')
# a=re.compile('a..z')
result = re.search(a, 'a9!z') 
print(result)
```

#### 3) *
- 별표(*) 바로 앞 문자가 0번 이상 반복
  - (예시 1) .* -> 하나 이상의 문자를 포함하는 문자열 지칭
  - (예시 2) ab*c -> b를 0번 이상 반복하는 문자 (ex. ac,az,a123c,abbbb)
  - (예시 3) like.* -> like에 하나 이상의 문자가 추가된 문자열 (ex. like, likely, likelihood)

#### 4) +
- 바로 앞 문자가 1번 이상 반복
  - (예시 1) ca+t -> a가 1개 이상 반복되어야 함 (ex. cat, caat, caaaat)
  - (예시 2) car+ot -> r이 1번 이상 반복되어야 함 (ex. carrot, carrrrrrrot)
  - (예시 3) like.+ -> like에 하나 이상의 문자열이 추가되어야 함 (ex. likeli, liker ; like는 안됨)
  - (예시 4) [A-Z]+ -> 대문자로만 이루어진 문자열 (ex. ABC, ABE, EFEF)

```python
#참고
example = "Luke Skywarker 02-123-4567 luke@daum.net" 
m1 = re.findall('\w', example) # 단어가 아니라 문자 단위로 출력 
m2 = re.findall('\w+', example) # 단어 단위로 출력
m3 = re.findall('\w*', example) # 단어 단위로 출력 

print("m1 : ", m1)
print("m2 : ", m2)
print("m3 : ", m3)
```

#### 5) ?
- 반복횟수 0~1회를 의미 (즉, 선행문자가 있어도되고 없어도 된다는 의미)
  - (예시) ab?c -> b가 있어도 되고 없어도 됨 -> abc 혹은 ac둘다 가능

#### 6) {}
- 바로 앞 문자의 반복횟수 지정
  - (예시 1) ca{2}t -> a가 2회 반복되어야 함 (ex. caat)
  - (예시 2) ca{2,5}t	-> a가 2회 이상 5회 이하 반복되어야 함 (ex. caat, caaat, caaaat)
  - (예시 3) ca{0, }t	-> 반복횟수 0회 이상 (*와 동일)	(ex. ct, cat, caat, caaat, caaaat)
  - (예시 4) cat{0, 1}t -> 반복횟수 0회 ~ 1회 이하 (?와 동일)	(ex. ct, cat)
  - (예시 5) cat{ , 3} -> 반복횟수 0회 이상 ~ 3회 이하	(ex. ct, cat, caat, caat)

```python
#참고
example = "ct cat caat caaat caaaat"
m1 = re.findall("ca{2}t", example)
m2 = re.findall("ca{2,5}t", example)
m3 = re.findall("ca{0,}t", example)
m4 = re.findall("ca{0,1}t", example)
m5 = re.findall("ca{,3}t", example)
print("m1 : ", m1)
print("m2 : ", m2)
print("m3 : ", m3)
print("m4 : ", m4)
print("m5 : ", m5)
```

#### 7) | 
- OR을 의미

```python
#예시
p = re.compile('a|b') # a 또는 b를 찾는다. 
m = p.findall("abcdefg")
print(m)
```

#### 8) ^ (시작) \$(끝)
- ^는 맨 처음에 어떤 것이 와야하는지 명시 
- $는 맨 마지막에 어떤 것이 와야하는지 명시

```python
# 메타문자 ^
m1 = re.findall("^Life", "Life is too short")
m2 = re.findall("^is", " is too short")
print("m1 결과 : ", m1)
print("m2 결과 : ", m2)
```

```python
m1 = re.findall("short$", "Life is too short")
m2 = re.findall("short$", "Life is too short. So what?")
 
print("m1 결과 : ", m1)
print("m2 결과 : ", m2)
```

#### 9) () 
- 그룹핑을 의미하며, 검색 결과의 특정 부분만 출력해주는 문자

```python
# 참고 
p = re.compile(r"\w+)\s+(\d+[-]\d+[-]\d+)")
m = p.search("park 010-1234-1234")
print(m.group(0))
print(m.group(1))
print(m.group(2))
```

#### 10) 기타
- [가-힣] : 완성형 한글
- r'정규표현식'
  - r은 Raw String의 약자로, r이 붙으면 순수한 문자임을 명시해주는 (백슬래쉬 2개를 쓰는 것과 동일한 역할을 해준다)
- (?:...)
  - 입력한 패턴값을 변수로 저장하지 않음
- (?=...)
  - ...에 해당되는 정규식과 매치되어야 함 (단, 해당 패턴은 리턴하지 않음)
- (?!...)
  - ...에 해당되는 정규식과 매치되지 말아야 함

```python
# (?=...) 활용
# 'http' 단어 추출희망 시, 아래와 같이 입력하면 https: 가 추출됨 
p = re.compile(".+:")
m = p.search("http://google.com")
print(m.group())
```

```python
# 위 상황에서 :를 제거하기 위해, 긍정형 전방 탐색 메타문자'(?!...)'를 활용
p = re.compile(".+(?=:)")
m = p.search("http://google.com")
print(m.group())
```

```python
# (?!...) 활용
# foo.bar, autoexec.bat, sendmail.cf, autoexec.exe과 같은 파일 확장명에서, 'bat'와,'exe' 형식을 제외하고 찾고 싶을 때
p = re.compile(".*[.].*$")
m = p.search("autoexec.bat")
print(m)
```

```python
p = re.compile(".*[.](?!bat$|exe$).*$") 
m = p.search("sendmail.cf")
print(m)
```

#### 10-2) groupdict()
- 패턴에 이름을 부여하였을 경우, groupdict 사용 시 딕셔너리 형태로 (이름:해당문자) 출력해줌 
- 패턴에 이름을 부여하기 위해서는, '?P<이름>' 형식으로 앞에 추가해주어야 함 

```python
result = re.match('(?P<지역번호>\d{2})-(?P<중간번호>\d{3,4})-(?P<뒷번호>\d{4})', '02-123-1234')
print(result.groupdict())
```

```python
print(result.groups())
```

```python
print(result.group(1))
```

```python
print(result.group('지역번호'))
```

```python
# step 2 ':'를 기준으로 구분
s = [re.split(":", entry, 2) for entry in m]
s
```

##### [예제 3] 한글만 골라내기

'韓子는 싫고, 한글은 nice하다. English 쵝오 -_-ㅋㅑㅋㅑ ./?!'

**(답)**

```python
s='韓子는 싫고, 한글은 nice하다. English 쵝오 -_-ㅋㅑㅋㅑ ./?!'

hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글과 띄어쓰기를 제외한 모든 글자
 
result = hangul.sub('', s) # 한글과 띄어쓰기를 제외한 모든 부분을 제거
print (result)
result = hangul.findall(s) # 정규식에 일치되는 부분을 리스트 형태로 저장
print(result)
```
